Using online stock trading data to do basic stock analysis and investment modeling

```python
import pandas as pd
import pandas_datareader.data as web   # Package and modules for importing data; this code may change depending on pandas version
import datetime
```

```python
# Setting start and end time for the market history
start = datetime.datetime(2016,1,1)
end = datetime.date.today()

symbols = array([
    'blv',  # bond: long-term bond
    'bnd',  # bond: total bond market
    'bsv',  # bond: short-term bond
    'tip',  # bond: treasury inflation protected
    'gld',  # commodity: gold (SPDR)
    'iau',  # commodity: gold (iShares)
    'slv',  # commodity: silver
    'vti',  # stock: US total stock market
    'vbk',  # stock: US small-cap stock
    'aapl', # stock: Apple
    'spy',  # index: S&P 500
          ])
# Get a stock historical data online
# First argument is the series we want, second is the source ("yahoo" for Yahoo! Finance), third is the start date, fourth is the end date
apple = web.DataReader("AAPL", "google", start, end)
gld = web.DataReader("GLD", "google", start, end)
spy = web.DataReader("SPY", "google", start, end)
bnd = web.DataReader("BND", "google", start, end)
all = web.DataReader(symbols, "google", start, end)
# all is a panel
all['Close'].head()
```

```python
all.dropna(inplace=True)
all['Close'].head()
```

```python
portfolios_symbols = ['bnd', 'spy', 'gld']
all['Close'][portfolios_symbols].plot()
```

```python
# Normalize the price based on the first day price
all.norm = all['Close'][portfolios_symbols].copy()
for s in portfolios_symbols:
    all.norm[s] = all['Close'][s]/all['Close'][s][0]
all.norm.plot()
```
