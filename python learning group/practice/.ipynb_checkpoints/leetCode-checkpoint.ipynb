## Q1 Coin Path
Given an array A (index starts at 1) consisting of N integers: $A_1, A_2, ..., A_N$ and an integer $B$. The integer $B$ denotes that from any place (suppose the index is $i$) in the array $A$, you can jump to any one of the place in the array $A$ indexed $i+1, i+2, …, i+B$ if this place can be jumped to. Also, if you step on the index i, you have to pay Ai coins. If $A_i$ is -1, it means you can’t jump to the place indexed i in the array.

Now, you start from the place indexed 1 in the array $A$, and your aim is to reach the place indexed N using the minimum coins. You need to return the path of indexes (starting from 1 to N) in the array you should take to get to the place indexed N using minimum coins.

If there are multiple paths with the same cost, return the lexicographically smallest such path.

If it's not possible to reach the place indexed N then you need to return an empty array.

```python
# Dynamic Programming
def cheapestMove(A, B):
    n = len(A)-1
    next = [n]
    dp = []
    result = []
    cost = 10000
    if(len(A)<=B):
        for i in range(n,0,-1):
            if(A[i]>=0):
                cost = A[-1] # last element
                next = [i]+next
                break
    else:
        for i in range(len(A)-1, len(A)-B, -1): 
            if(A[i]>=0):
                #print(A[0:i])
                r = cheapestMove(A[0:i], B)
                #print(r)
                newCost = A[n]+r['cost']
                if(newCost<=cost):
                    cost = newCost
                    next = [r['next']]+next
    return{'cost': cost, 'next': next}
        
A, B = [1,2,4,-1,2], 2
res = cheapestMove(A, B)
print(res)
```
